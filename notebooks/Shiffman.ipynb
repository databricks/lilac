```python
# Read all json files in /Users/nikhil/Downloads/thesis-data/*.json

import json
import os
import sys


DIR = '/Users/nikhil/Downloads/thesis-data/'


# List all the json files in DIR.
def list_json_files():
  json_files = []
  for file in os.listdir(DIR):
    if file.endswith('.json'):
      json_files.append(file)
  return json_files


json_files = list_json_files()

project_outputs = []
for file in json_files:
  with open(DIR + file) as f:
    file_json = json.load(f)
  project_outputs.extend(file_json['projects'])

print(len(project_outputs))

for i in range(10):
  print(project_outputs[i])

# Write project_outputs to a jsonl file.
with open(DIR + 'all_projects.jsonl', 'w') as f:
  for project in project_outputs:
    f.write(json.dumps(project) + '\n')
```
